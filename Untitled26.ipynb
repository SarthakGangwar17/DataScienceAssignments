Q1. Why use the Exception class for custom exceptions:

Inheritance and Handling: Custom exceptions inherit from Exception to become part of the exception hierarchy. This allows them to be caught and handled using try...except blocks like any other built-in exception.
Organization and Clarity: By inheriting from Exception, custom exceptions gain structure and can be grouped with related exceptions for better code organization and understanding.
Custom Attributes: You can add custom attributes to your exceptions to store additional error-related information, providing richer context for handling and debugging.
Consistency and Compatibility: Using Exception as the base class ensures consistency with Python's exception handling mechanisms and compatibility with libraries and frameworks that expect exceptions to follow this structure.

```python
import pprint

def print_exception_hierarchy(base_class=Exception):
    subclasses = base_class.__subclasses__()
    pprint.pprint(subclasses)
    for subclass in subclasses:
        print_exception_hierarchy(subclass)

print("Python Exception Hierarchy:")
a=print_exception_hierarchy()

```
