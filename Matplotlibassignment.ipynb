Answer 2:-
Scatter = It is plot in matplotlib that is used to represent the relation between the x axis and y axis. It is done with the help of dots.


```python
import numpy as np
import matplotlib.pyplot as plt


x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
colr=np.arange(1,51);

plt.scatter(x,y,c=colr)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is scatter plot")
```

It is function in matplotlib by which
we can create multiple plot in single figure



```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("first plot")


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Second Plot")


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Third plot")


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Fourth plot")
```

Answer 4:-
Bar plot is a type of plot in matplotlib that is basically used to represent the categorical cata
and continous data.



```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.title("Profit")
plt.xlabel("Company")
plt.ylabel("total profit")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add labels and title
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```
